```python
import seaborn as sns
import matplotlib.pyplot as pt
import pandas as pd
```

```python
#load dataset
tips=sns.load_dataset('tips')
print(tips.head())
```

```python
tips.info()
```

```python
tips.describe()
```

```python
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='time')
pt.title('Total_Bill vs Tip')
pt.xlabel('Total_Bill')
pt.ylabel('Tip')
pt.show()

```

```python
sns.lineplot(data=tips, x='total_bill', y='tip', hue='time')
pt.title('Total_Bill vs Tip')
pt.xlabel('Total_Bill')
pt.ylabel('Tip')
pt.show()
```

```python
fmri=sns.load_dataset('fmri')
print(fmri.head())
```

```python
sns.lineplot(data=fmri, x='timepoint', y='signal', hue='event', style='region')
pt.title('FMRI Signal Over Time')
pt.xlabel('Timepoint')
pt.ylabel('Singal')
pt.show()
```

```python
sns.barplot(data=tips,x='day',y='total_bill',hue='sex')
pt.title=('Average Total_bill Per Day by Sex')
pt.xlabel('Day')
pt.ylabel('AverageTotal_Bill')
pt.show()
```

```python
sns.histplot(data=tips,x='total_bill', bins=10,kde=True)
pt.title('Distribution od Total_Bill')
pt.xlabel('Total_Bill')
pt.ylabel('Frequency')
pt.show()
```

```python
sns.boxplot(data=tips, x='day',y='total_bill',hue='smoker')
pt.title=('Boxplot of Total_Bill by Day and Smoker Status')
pt.xlabel('Day')
pt.ylabel('Total_Bill')
pt.show()
```

```python
iris=sns.load_dataset('iris')
print(iris.head())
```

```python
sns.pairplot(iris,hue='species')
pt.suptitle('Iris Pairplot',y='1.02')
pt.show()
```

```python
violinplot=sns.violinplot(data=iris,x='species',y='sepal_length',inner='quartile')
pt.title('Count of Each Iris Species')
pt.xlaybel('Species')
pt.ylabel('Sepal_Length')
pt.show()
```
